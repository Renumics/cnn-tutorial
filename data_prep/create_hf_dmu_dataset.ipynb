```python
from renumics import spotlight

# Load the dataset
meshes = []
ids = []
labels = []
split = []

with spotlight.Dataset("../datasets/dmu_dataset_base.h5", "r") as dataset:
    print(len(dataset))
    print(sorted(dataset.keys()))
    for row in dataset:     
        meshes.append(row['mesh'].encode().tolist())
        ids.append(row['id'])
        labels.append(row['label'])
        split.append(row['split'])
        
```

```python
import pandas as pd

df = pd.DataFrame({'mesh': meshes, 'id': ids, 'split': split})
df_test = df[df['split'] == 'test']
df_train = df[df['split'] == 'train']
```

```python
#add 3d meshes and voxels
import numpy as np

class_names = ['Nut', 'Screw', 'GearWheel']
data = np.load('../datasets/data_res48-48-48.npz')
train_geometries, train_labels = data['train_geometries'], data['train_labels']
test_geometries, test_labels, test_ids = data['test_geometries'], data['test_labels'], data['test_ids']

```

```python
from skimage import measure
from trimesh import voxel
import trimesh

meshes_voxelized = []
voxels_test = []


for i in range(test_geometries.shape[0]):
    voxel_geo = test_geometries[i,:,:,:]
    #verts, faces, normals, values = measure.marching_cubes(voxel, step_size=1)
    voxel_grid=voxel.VoxelGrid(voxel_geo)
    voxelized_mesh = voxel_grid.as_boxes()
    mesh_voxelized = spotlight.Mesh(voxelized_mesh.vertices, voxelized_mesh.faces)
    voxels_test.append(voxel_geo)
    meshes_voxelized.append(mesh_voxelized.encode().tolist())

#df_test['voxels']=voxels
df_test['mesh_voxelized']=meshes_voxelized



```

```python
meshes_voxelized = []
voxels_train = []



for i in range(train_geometries.shape[0]):
    voxel_geo = train_geometries[i,:,:,:]
    #verts, faces, normals, values = measure.marching_cubes(voxel, step_size=1)
    voxel_grid=voxel.VoxelGrid(voxel_geo)
    voxelized_mesh = voxel_grid.as_boxes()
    mesh_voxelized = spotlight.Mesh(voxelized_mesh.vertices, voxelized_mesh.faces)
    voxels_train.append(voxel_geo)
    meshes_voxelized.append(mesh_voxelized.encode().tolist())

#df_train['voxels']=voxels
df_train['mesh_voxelized']=meshes_voxelized
```

```python
import datasets

# Create the dataset
class_names=['Nut', 'Screw', 'GearWheel']
ds_test = datasets.Dataset.from_pandas(df_test, split='test')
ds_test_voxels = datasets.Dataset.from_dict({'voxel': voxels_test, 'label': test_labels},
                                            datasets.Features({'voxel': datasets.Array3D(shape=(48,48,48),dtype='float32'), 'label': datasets.ClassLabel(names=class_names)}))
ds_test= datasets.concatenate_datasets([ds_test, ds_test_voxels],axis=1)

ds_train = datasets.Dataset.from_pandas(df_train, split='test')
ds_train_voxels = datasets.Dataset.from_dict({'voxel': voxels_train, 'label': train_labels},
                                            datasets.Features({'voxel': datasets.Array3D(shape=(48,48,48),dtype='float32'), 'label': datasets.ClassLabel(names=class_names)}))
ds_train= datasets.concatenate_datasets([ds_train, ds_train_voxels],axis=1)

ds = datasets.DatasetDict({'train': ds_train, 'test': ds_test})

```
